# Question 2 (20 Points) - Functions



- ABC corp. is considering a potential project, which is associated with a constant growing perpetuity cash flow pattern described below. (If you cannot remeber what is constant growing perpetuity, check out this link: https://studyfinance.com/present-value-of-growing-perpetuity/)

\begin{equation}
  \text{$V_{Project}$} =  
    \begin{cases}
       0 & \text{if g<=0}\\
       \dfrac{CF}{(r-g)} & \text{if 0<g<r} \\
       0 & \text{if  g > r}
    \end{cases}       
\end{equation}

1. (14 points) Please define a function with conditional statements embeded to fully capture the value of the project, given the inputs `CF`, `r`, and `g`
    - with your defined function, user should see the returned value along with the following explaination.
        - (1) If g<=0, you should print string 'Abandon the Project' along with the returned value; 
        - (2) if 0<g<r, you need to print string 'The model is Valid with a non-zero PV' along with the returned value;
        - (3) if g is greater than r, you should print string 'g is greater than the required return, model breaks'
    
2.  (6 points) Given the following inputs, please ***use the function you just defined*** to establish the value of this asset 
    - Input set $f(CF, f, g)$\#1: CF = 1000, r=0.05, g = -0.02
    - Input set $f(CF, f, g)$\#2: CF = 1000, r=0.08, g = 0.03

___Hint:___
- you can literally eyeball the answer(s)  
- Be careful about the ***indentation***


```python
def calculate_project_value(CF, r, g):

    # Using a nested-if function, I input the parameters

    if g <= 0:
        # If g is less than or equal to 0: abandon
        print("Abandon the Project")
        return 0
    elif 0 < g < r:
        # If between
        PV = CF / (r - g)
        print("The model is Valid with a non-zero PV")
        return PV
    elif g >= r:
        # If g is greater than or equal to r, the model breaks
        print("g is greater than the required return, model breaks")
        return None

# Input set #1: CF = 1000, r = 0.05, g = -0.02
value1 = calculate_project_value(1000, 0.05, -0.02)
print("Value of the Project for Input Set #1:", value1)

# Input set #2: CF = 1000, r = 0.08, g = 0.03
value2 = calculate_project_value(1000, 0.08, 0.03)
print("Value of the Project for Input Set #2:", value2)
```

# Weighted Portfolio return with  the given weights

```python
stocks = ['NFLX', 'KO', 'AMZN', 'META', 'LUV', 'TSLA']
weights = np.array([0.15, 0.10, 0.35, 0.23, 0.07, 0.10])
stock_data = yf.download(stocks, start='2016-01-01', end='2023-12-08')

# First, I calculate the weighted log returns each stock

log_returns = np.log(stock_data['Close'] / stock_data['Close'].shift(1))
weighted_log_returns = log_returns.mul(weights, axis=1)

# Then, portfolio cumulative return

portfolio_log_return = weighted_log_returns.sum(axis=1)
cumulative_portfolio_return = np.exp(portfolio_log_return.cumsum()) - 1
print("Cumulative Portfolio Return:")
print(cumulative_portfolio_return)
```

# Estimate the portfolio beta,  𝛽𝑝𝑜𝑟𝑡𝑓𝑜𝑙𝑖𝑜  in the spirit of the Capital Asset Pricing Model 
# (assuming risk-free rate is close to zero, for simplicity)
# Let's use the return on the SPDR S&P 500 ETF as a proxy for market return.

```python
# Beta is the co-variance divided by the variance of the market
stocks = ['NFLX', 'KO', 'AMZN', 'META', 'LUV', 'TSLA']
weights = np.array([0.15, 0.10, 0.35, 0.23, 0.07, 0.10])

# Importing the data of the SPY from Yahoo Finance
all_data = yf.download(stocks + ['SPY'], start='2016-01-01', end='2023-12-08')['Close']
daily_returns = all_data.pct_change().dropna()
portfolio_returns = (daily_returns[stocks] * weights).sum(axis=1)

# Covariance and Variance gives the Beta
covariance = np.cov(portfolio_returns, daily_returns['SPY'])[0][1]
market_variance = daily_returns['SPY'].var()
portfolio_beta = covariance / market_variance
print("The Portfolio Beta is:", portfolio_beta)
```

# Question 4 (25 Points) - Capital Budgeting

- Context: Your company is considering a new capital investment of a large equipment. The future project cash flows associatated with this investment is `[5300, 5400, 5895, 6000, 7008, 7532, 8000, 8032, 8521, 8600, 9023]` (starting from $t_1$). The risk-adjusted discount rates for each invetment period are estimated as `[0.101, 0.092, 0.112, 0.113, 0.130, 0.145, 0.145, 0.149, 0.113, 0.151, 0.156] `(starting from $t_1$).
    - Determine the number of investment periods `T`.
    - (10 points) What is the net present value (NPV) of this project? Given the initial equipment purchasing price to be $\$15038$ (at $t_0$) and the initial consulting fee of $\$5355$ (at $t_0$).
    - (5 points) Given the above project cashflow structure, what is the internal rate of return (IRR)?
    - (4 points)  If the discount rates vector uniformly decreases by  -0.004, what is the new NPV generated by this investment? What is the percentage change (sensitivity) in NPV ($\% \dfrac{\Delta NPV}{NPV}$) associated with this discount rates change? 
    - (6 points)  Discuss if your project CF is further assumed to be growing over time, the growth rate $g$ is subject to a normal distribution $N(\mu=0.0035, \sigma=0.04)$, the first stream of project cash flow ($CF_{t_1}$) is $\$5300$, ___describe___ in detail how to model your project PCF

```python
# Given cash flows and discount rates
cash_flows = [5300, 5400, 5895, 6000, 7008, 7532, 8000, 8032, 8521, 8600, 9023]
discount_rates = [0.101, 0.092, 0.112, 0.113, 0.130, 0.145, 0.145, 0.149, 0.113, 0.151, 0.156]

# Investment costs
initial_equipment_cost = 15038
initial_consulting_fee = 5355
initial_investment = initial_equipment_cost + initial_consulting_fee

T = len(cash_flows) #Len counts the number of cash flow periods above
```

```python
!pip install numpy-financial
```

# Calculate the NPV and IRR

```python
# Calculating NPV

npv = -initial_investment
for t in range(T):
    npv += cash_flows[t] / (1 + discount_rates[t]) ** (t + 1)

# Next, Calculating the IRR

import numpy_financial as npf
irr = npf.irr([-initial_investment] + cash_flows)
print(f"Number of Investment Periods (T): {T}")
print(f"Net Present Value (NPV): ${npv:.2f}")
print(f"Internal Rate of Return (IRR): {irr:.4f}")
```

# Calculate the NEW NPV if discount rate decreases by -0.004

```python
# Decreasing the discount rate by 0.004
adjusted_discount_rates = [rate - 0.004 for rate in discount_rates]

# Calculating the new NPV
new_npv = -initial_investment
for t in range(T):
    new_npv += cash_flows[t] / (1 + adjusted_discount_rates[t]) ** (t + 1)

print(f"New Net Present Value (NPV) with Adjusted Discount Rates: ${new_npv:.2f}")
```

# Sensitivity of NPV to the new discount rate

```python
# I calculated the percentage change
percentage_change_npv = ((new_npv - npv) / npv) * 100
print(f"Percentage Change in NPV: {percentage_change_npv:.2f}%")
```

```python
# Customer visits data for each store

store_1_visits = [1169, 1245, 1400, 1305, 2002, 3020, 4540, 3987, 4900, 5010, 6939, 8989]
store_2_visits = [780, 804, 930, 1305, 1504, 2829, 3542, 4001, 3764, 4703, 6523, 8010]
store_3_visits = [1552, 1621, 1830, 1902, 2098, 3923, 5563, 5017, 6102, 6880, 7069, 9140]
```

```python
# Customer flow is calculated as monthly visit/total visits

def calculate_percentages(visits):
    total_annual_visits = sum(visits)
    return [(month_visits / total_annual_visits) * 100 for month_visits in visits] #this gives percentages

# Monthly Percentages for each store
store_1_percentages = calculate_percentages(store_1_visits)
print("Store 1 Monthly Percentages:", store_1_percentages)

store_2_percentages = calculate_percentages(store_2_visits)
print("Store 2 Monthly Percentages:", store_2_percentages)

store_3_percentages = calculate_percentages(store_3_visits)
print("Store 3 Monthly Percentages:", store_3_percentages)
```
